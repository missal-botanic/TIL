```python
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv('LinearRegressionData.csv')

X = dataset.iloc[:, :-1].values #마지막 제외하고
y = dataset.iloc[:, -1].values #마지막 하나

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)

y_pred = reg.predict(X_train)
```

```python
plt.scatter(X_train ,y_train, color ='blue')
plt.plot(X_train, y_pred, color = 'green')
plt.title('Score by hour Train')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()

plt.scatter(X_test ,y_test, color ='blue')
plt.plot(X_train, y_pred, color = 'green') # 그대로 두기
plt.title('Score by hour Test')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()

print('9시간 공부 했을 때', reg.predict([[9]]))
print(reg.coef_) #기울기
print(reg.intercept_) # y절편
print(f"모델 방정식: y = {reg.coef_[0]} * X + {reg.intercept_}")
```

```python
print(reg.score(X_test, y_test))
print(reg.score(X_train, y_train))
```

```python
y_pred_df = pd.DataFrame(y_pred, columns=['Predicted'])
dataset['Predicted'] = y_pred_df
print(dataset)
```
