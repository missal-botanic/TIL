```python
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv('LinearRegressionData.csv')

X = dataset.iloc[:, :-1].values #마지막 제외하고
y = dataset.iloc[:, -1].values #마지막 하나

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import SGDRegressor 
sr = SGDRegressor(max_iter=150, eta0=0.0001, random_state=0, verbose=1)
sr.fit(X_train, y_train)
```

```python
plt.scatter(X_train ,y_train, color ='blue')
plt.plot(X_train, sr.predict(X_train), color = 'green')
plt.title('Score by hour SGD')
plt.xlabel('hours')
plt.ylabel('score')
plt.show()
print(sr.coef_) #기울기
print(sr.intercept_) # y절편
print(sr.score(X_test, y_test))
print(sr.score(X_train, y_train))
```
