```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split as tts
from keras.models import Sequential as SL
from keras.layers import Dense
from keras.layers import Dropout

data = np.array([
    [1, 2, 0],
    [2, 1, 0],
    [3, 1, 0],
    [1, 3, 0],
    [2, 2, 0],
    [7, 7, 1],
    [6, 8, 1],
    [8, 6, 1],
    [7, 6, 1],
    [6, 7, 1]
])

# X와 Y를 분리
X = data[:, 0:2]  # X는 첫 두 열 (x, y)
Y = data[:, 2]    # Y는 세 번째 열 (z)

# train-test split
X_tr, X_te, y_tr, y_te = tts(X, Y, test_size=0.2, random_state=42)

# 모델 정의
model = SL()
model.add(Dense(5, activation="relu"))
model.add(Dense(100, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

model.add(Dropout(0.1)) #10%

model.compile(loss="mean_squared_error", metrics=["accuracy"])

# 모델 학습
model.fit(X_tr, y_tr, epochs=100, batch_size=3, verbose=1)

# 모델 평가
loss, accuracy = model.evaluate(X_te, y_te)

# 모델 요약
model.summary()

print(f"Loss: {loss}, Accuracy: {accuracy}")
```
